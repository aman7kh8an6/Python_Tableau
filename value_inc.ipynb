```python
import pandas as pd

data = pd.read_csv('/home/aman/Documents/Data Analysis/Value_Inc_Sales/transaction.csv', sep = ';')

costPerItem = data["CostPerItem"]
sellingPricePerItem = data["SellingPricePerItem"]
noOfItemPurchased = data["NumberOfItemsPurchased"]
profitPerItem = sellingPricePerItem - costPerItem

profitPerTransacion = noOfItemPurchased*profitPerItem
costPerTransaction = noOfItemPurchased*costPerItem
sellingPricePerTransaction = noOfItemPurchased*sellingPricePerItem

data["CostPerTransaction"] = costPerTransaction 
data["SellingPricePerTransaction"] = sellingPricePerTransaction
data["ProfitPerTransacion"] = profitPerTransacion
data["Markup"] = round(data["ProfitPerTransacion"]/costPerTransaction, 2)

date = data['Day'].astype(str) + '-' + data['Month'].astype(str) + '-' + data['Year'].astype(str)
data["Date"] = date

split_clientKeywords = data["ClientKeywords"].str.split(',', expand = True)
data["ClientAge"] = split_clientKeywords[0]
data["ClientType"] = split_clientKeywords[1]
data["LengthOfContract"] = split_clientKeywords[2]
data["ClientAge"] = data["ClientAge"].str.replace('[','')
data["LengthOfContract"] = data["LengthOfContract"].str.replace(']','')

data["ItemDescription"] = data["ItemDescription"].str.lower()
 
seasons = pd.read_csv('/home/aman/Documents/Data Analysis/Value_Inc_Sales/value_inc_seasons.csv', sep = ';')

data = pd.merge(data, seasons, on = 'Month')
data = data.drop(['Day', 'Month','Year','ClientKeywords'], axis = 1)

data.to_csv('/home/aman/Documents/Data Analysis/Value_Inc_Sales/valueInc_cleaned.csv', index = False)
data
```
